## Chapter 1: Essential building blocks

````python
import matplotlib.pyplot as plt
import numpy as np

# figure and axes object,
# (nrows, ncols): 	Number of rows/columns of subplots (default is 1x1 )
fig, ax = plt.subplots()

x = np.linspace(0,10,100) # 100 points from 0 to 10 inclusively.
y = np.sin(x)

ax.plot(x,y, linewidth=2, label='sin(x)') # label won't show until, we call ax.legend()

# Step 4: Customize the plot
ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('Y values', fontsize=12)
ax.set_title('My First Professional Plot', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Step 5: Display
plt.tight_layout()
plt.show()
````

##  Essential concepts: 
### 1. color marker linestyle:
- **Color** (e.g. 'b' = blue)

- **Marker** (e.g. 'o' = circle at each data point)

- **Line style** (e.g. '-' = solid line)

---
### 1. Colors:
You can use either a **single-letter shorthand** or **full names**:
| Shorthand | Color   |
| --------- | ------- |
| `'b'`     | blue    |
| `'g'`     | green   |
| `'r'`     | red     |
| `'c'`     | cyan    |
| `'m'`     | magenta |
| `'y'`     | yellow  |
| `'k'`     | black   |
| `'w'`     | white   |

### 2. Markers
Markers are used to represent individual data points.
| Symbol | Marker         |
| ------ | -------------- |
| `'o'`  | circle         |
| `'^'`  | triangle\_up   |
| `'v'`  | triangle\_down |
| `'s'`  | square         |
| `'p'`  | pentagon       |
| `'*'`  | star           |
| `'+'`  | plus sign      |
| `'x'`  | x              |
| `'D'`  | diamond        |


### 3. Line Styles
These define how the line connecting the points looks

| Symbol | Style                  |
| ------ | ---------------------- |
| `'-'`  | solid line             |
| `'--'` | dashed line            |
| `'-.'` | dash-dot line          |
| `':'`  | dotted line            |
| `''`   | no line (just markers) |

## 1. Line plots
- **ax.plot()** or **plt.plot()**

#### Parameters: 

| Parameter            | Description                                                  |
| -------------------- | ------------------------------------------------------------ |
| `x`                  | Sequence of x-values                                         |
| `y`                  | Sequence of y-values                                         |
| `format_string`      | Optional shorthand like `'ro--'` (color, marker, line style) |
| `color`              | Line color: `'r'`, `'g'`, `'b'`, `'k'`, etc.                 |
| `linestyle`          | Line style: `'-'`, `'--'`, `'-.'`, `':'`, `'None'`           |
| `marker`             | Point marker: `'o'`, `'+'`, `'*'`, `'.'`, `'x'`, etc.        |
| `label`              | Label for legend                                             |
| `linewidth` or `lw`  | Width of the line                                            |
| `markersize` or `ms` | Size of the markers                                          |
| `alpha`              | Transparency (0 to 1)                                        |
| `zorder`             | Drawing order: higher is on top                              |


### NOTE : Positional Arguments (in order)
- ***These must be passed without a keyword***.
- And ***order*** must also be followed.

    1. **x:** array-like (e.g., list or NumPy array)

    2. **y:** array-like (same length as x)

    3. **format_string** (optional): A string combining:

        - Color (e.g. 'r', 'g')

        - Marker (e.g. 'o', '+')

        - Line style (e.g. '-', '--')


````python
# Line plot using `Global State`

import matplotlib.pyplot as plt

months = [1,2,3,4]
sales =  [1000, 1500, 1200, 1800]

# plt.plot(months, sales, linestyle="--", marker="o", linewidth=2, label="sales data", color="Blue")

# Shortcut:
plt.plot(months, sales, "bo--" ,linewidth=2, label="sales data", markersize=10)

# labels  X and Y
plt.xlabel("Months", fontsize=14, fontstyle='normal', fontweight='bold')
plt.ylabel("Sales Per Month", fontsize=14)
plt.legend(loc="upper left", fontsize=12)
plt.grid(alpha=0.2, color='r', linestyle='--', lw=1)
plt.xlim(0,5)
plt.ylim(0,2000)


plt.tight_layout()
plt.show()
````

````python
# Training curve example

import matplotlib.pyplot as plt

epochs = list(range(1, 11))
train_loss = [0.8, 0.6, 0.45, 0.35, 0.28, 0.22, 0.18, 0.15, 0.13, 0.12]
val_loss = [0.82, 0.63, 0.48, 0.41, 0.38, 0.36, 0.35, 0.34, 0.35, 0.36]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(epochs,train_loss, "ro-", label="Training Loss" )
ax.plot(epochs,val_loss, "bD-", label="Value Loss")

ax.set_xlabel("X")
ax.set_ylabel("Y")

ax.set_title("Line Plot Demonstration")
ax.legend() # labels won't show until, it is called
ax.grid()
ax.set_in_layout(True)
````

## 2. Scatter plot:

`scatter(x, y, s, c, marker, cmap, norm, vmin, vmax, alpha, linewidths, edgecolors, colorizer, plotnonfinite, data, **kwargs)`

````python
from sklearn.datasets import make_classification

# Generate sample classification data
X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, 
                          n_informative=2, n_clusters_per_class=1, 
                          random_state=42)

# X = list, containing two features, and y is target variable

# Let's plot a scatter plot to define relationship between features, this plot will showcase:
# 1. Relationship b/w features where y=0,
# 2. Relationship b/w features where y=1

colors = ["blue", "Coral"]

for i, color in enumerate(colors):
    plt.scatter(X[y==i, 0], X[y==i, 1], c=color, label=f"class {i}", s=20)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Relationship between features")
plt.tight_layout()
plt.legend()


````

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 3.5]
y = [5, 7, 6, 8, 7]
sizes = [100, 200, 300, 400, 500]
colors = [10, 20, 30, 40, 50]

plt.scatter(x, y, 
            c=colors,        # color by value
            s=sizes,         # size of dots
            alpha=0.6,       # transparency
            cmap='plasma',   # color map
            edgecolors='black',
            linewidths=5,
            marker='o',
            label='Points')

plt.colorbar(label='Color Scale')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Custom Scatter Plot")
plt.legend()
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5, 6]
scores = [50, 55, 65, 70, 75, 80]
stress = [30, 20, 40, 35, 25, 10]
sleep = [6, 7, 5, 6, 8, 9]

plt.scatter(hours, scores, c=stress, s=[s*10 for s in sleep], cmap='coolwarm', alpha=0.8)
plt.colorbar(label='Stress Level')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("4D Scatter Plot")
plt.grid(True)
plt.show()
````

### HOW ABOVE PLOT IS 4-D ?
```py 
plt.scatter(hours, scores, c=stress, s=[s*10 for s in sleep], cmap='coolwarm', alpha=0.8)
```
 - This line is plotting a 4D dataset on a 2D graph.
 - Mathematically, we can only directly plot two axes (x and y). But we can encode two more dimensions visually using:
    - Color
    - Size

| Dimension  | Variable | Mapped To                     | Visual Meaning                          |
| ---------- | -------- | ----------------------------- | --------------------------------------- |
| 1️⃣ X-axis | `hours`  | `x` coordinate                | Hours studied                           |
| 2️⃣ Y-axis | `scores` | `y` coordinate                | Exam scores                             |
| 3️⃣ Color  | `stress` | `c=stress`, `cmap='coolwarm'` | Level of stress (shown via color scale) |
| 4️⃣ Size   | `sleep`  | `s=[s*10 for s in sleep]`     | Amount of sleep (bubble size)           |

### NOTE: `subplot()` and `subplots()` are two different functions.

````python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,15,11,18]

plt.subplot(1,2,1) # 1 row, 2 column, 1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)  # 1 row, 2 column, 2nd subplot
plt.bar(x,y)
plt.title("Bar Chart")

plt. tight_layout()
````

````python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,15,11,18]

fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(x,y)
ax[0].set_title("Line Plot")

ax[1].bar(x,y)
ax[1].set_title("Bar Plot")

# Title of whole fig
fig.suptitle("Comparision of Line and Bar plot")

plt.tight_layout()
````

## Understanding `rcParams` - Global settings

- `rcParams` is a dictionary-like object in **Matplotlib**.

- It controls the default behavior and appearance of all plots **globally**.


````python
import matplotlib.pyplot as plt

# View a setting
print("Size Before : ", plt.rcParams['font.size'])  # Current default font size

# Update global settings
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['axes.linewidth'] = 1.2

print("Size After : ", plt.rcParams['font.size'])  
````
