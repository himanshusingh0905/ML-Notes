## Chapter 1: Essential building blocks

````python
import matplotlib.pyplot as plt
import numpy as np

# figure and axes object,
# (nrows, ncols): 	Number of rows/columns of subplots (default is 1x1 )
fig, ax = plt.subplots()

x = np.linspace(0,10,100) # 100 points from 0 to 10 inclusively.
y = np.sin(x)

ax.plot(x,y, linewidth=2, label='sin(x)') # label won't show until, we call ax.legend()

# Step 4: Customize the plot
ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('Y values', fontsize=12)
ax.set_title('My First Professional Plot', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Step 5: Display
plt.tight_layout()
plt.show()
````

##  Essential concepts: 
### 1. color marker linestyle:
- **Color** (e.g. 'b' = blue)

- **Marker** (e.g. 'o' = circle at each data point)

- **Line style** (e.g. '-' = solid line)

---
### 1. Colors:
You can use either a **single-letter shorthand** or **full names**:
| Shorthand | Color   |
| --------- | ------- |
| `'b'`     | blue    |
| `'g'`     | green   |
| `'r'`     | red     |
| `'c'`     | cyan    |
| `'m'`     | magenta |
| `'y'`     | yellow  |
| `'k'`     | black   |
| `'w'`     | white   |

### 2. Markers
Markers are used to represent individual data points.
| Symbol | Marker         |
| ------ | -------------- |
| `'o'`  | circle         |
| `'^'`  | triangle\_up   |
| `'v'`  | triangle\_down |
| `'s'`  | square         |
| `'p'`  | pentagon       |
| `'*'`  | star           |
| `'+'`  | plus sign      |
| `'x'`  | x              |
| `'D'`  | diamond        |


### 3. Line Styles
These define how the line connecting the points looks

| Symbol | Style                  |
| ------ | ---------------------- |
| `'-'`  | solid line             |
| `'--'` | dashed line            |
| `'-.'` | dash-dot line          |
| `':'`  | dotted line            |
| `''`   | no line (just markers) |

---


## Optional: You can use kwargs instead
Instead of 'b-o', you can do:
```py
ax.plot(x, y, color='blue', linestyle='-', marker='o')
```

## 1. Line plots
- **ax.plot()** or **plt.plot()**

#### Parameters: 

| Parameter            | Description                                                  |
| -------------------- | ------------------------------------------------------------ |
| `x`                  | Sequence of x-values                                         |
| `y`                  | Sequence of y-values                                         |
| `format_string`      | Optional shorthand like `'ro--'` (color, marker, line style) |
| `color`              | Line color: `'r'`, `'g'`, `'b'`, `'k'`, etc.                 |
| `linestyle`          | Line style: `'-'`, `'--'`, `'-.'`, `':'`, `'None'`           |
| `marker`             | Point marker: `'o'`, `'+'`, `'*'`, `'.'`, `'x'`, etc.        |
| `label`              | Label for legend                                             |
| `linewidth` or `lw`  | Width of the line                                            |
| `markersize` or `ms` | Size of the markers                                          |
| `alpha`              | Transparency (0 to 1)                                        |
| `zorder`             | Drawing order: higher is on top                              |


### Positional Arguments (in order)
- ***These must be passed without a keyword***.
- And ***order*** must also be followed.

    1. **x:** array-like (e.g., list or NumPy array)

    2. **y:** array-like (same length as x)

    3. **format_string** (optional): A string combining:

        - Color (e.g. 'r', 'g')

        - Marker (e.g. 'o', '+')

        - Line style (e.g. '-', '--')


````python
# Line plot using `Global State`

import matplotlib.pyplot as plt

x = [1,2,3,4]
y =  [ 2 * i for i in x]

plt.plot(x, y, "ro-", label="Simple Plot", ms=10, lw=2)

plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Training curve example

import matplotlib.pyplot as plt

epochs = list(range(1, 11))
train_loss = [0.8, 0.6, 0.45, 0.35, 0.28, 0.22, 0.18, 0.15, 0.13, 0.12]
val_loss = [0.82, 0.63, 0.48, 0.41, 0.38, 0.36, 0.35, 0.34, 0.35, 0.36]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(epochs,train_loss, "ro-", label="Training Loss" )
ax.plot(epochs,val_loss, "bD-", label="Value Loss")

ax.set_xlabel("X")
ax.set_ylabel("Y")

ax.set_title("Line Plot Demonstration")
ax.legend() # labels won't show until, it is called
ax.grid()
ax.set_in_layout(True)
````

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

# Top left: Model accuracy comparison
models = ['Linear', 'Random Forest', 'SVM', 'Neural Net']
accuracy = [0.85, 0.92, 0.88, 0.94]
ax1.bar(models, accuracy, color=['blue', 'green', 'red', 'orange'])
ax1.set_title('Model Accuracy Comparison')
ax1.set_ylabel('Accuracy')

# Top right: Training curves
epochs = range(1, 11)
loss_a = [0.8, 0.6, 0.4, 0.3, 0.25, 0.22, 0.2, 0.18, 0.17, 0.16]
loss_b = [0.9, 0.7, 0.5, 0.35, 0.28, 0.25, 0.23, 0.22, 0.21, 0.20]
ax2.plot(epochs, loss_a, 'b-o', label='Model A')
ax2.plot(epochs, loss_b, 'r-s', label='Model B')
ax2.set_title('Training Loss Comparison')
ax2.legend()

# Bottom left: Data distribution
data = np.random.normal(50, 15, 1000)
ax3.hist(data, bins=30, alpha=0.7, color='skyblue')
ax3.set_title('Feature Distribution')

# Bottom right: Correlation matrix
corr_data = np.array([[1.0, 0.8, -0.3], [0.8, 1.0, -0.1], [-0.3, -0.1, 1.0]])
im = ax4.imshow(corr_data, cmap='coolwarm', aspect='auto')
ax4.set_title('Feature Correlation')

plt.tight_layout()
plt.show()
````
