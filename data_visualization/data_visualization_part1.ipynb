## standard deviation and variance

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(0)

# Generate three sets with different means
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=3, scale=1, size=1000)
data3 = np.random.normal(loc=-3, scale=1, size=1000)

plt.hist(data1, bins=30, alpha=0.5, label='loc=0')
plt.hist(data2, bins=30, alpha=0.5, label='loc=3')
plt.hist(data3, bins=30, alpha=0.5, label='loc=-3')
plt.legend()
plt.title("Effect of changing loc (mean)")
plt.show()
```

```python
# Generate three sets with different spreads
data1 = np.random.normal(loc=0, scale=0.5, size=1000)
data2 = np.random.normal(loc=0, scale=1, size=1000)
data3 = np.random.normal(loc=0, scale=2, size=1000)
data4 = np.random.normal(loc=0, scale=0.1, size=1000)

plt.hist(data1, bins=30, alpha=0.5, label='scale=0.5')
plt.hist(data2, bins=30, alpha=0.5, label='scale=1')
plt.hist(data3, bins=30, alpha=0.5, label='scale=2')
plt.hist(data4, bins=30, alpha=0.5, label='scale=0.5')

plt.legend()
plt.title("Effect of changing scale (standard deviation)")
plt.show()
```
