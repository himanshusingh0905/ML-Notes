```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Let's customize above plots:

plt.plot(x,y, color='red', linestyle='--', marker='o',linewidth=5, markersize=15)
plt.grid()
plt.xlabel('X - axis') 
plt.ylabel('Y - axis')  
plt.show()
```

```python
# Multiple plots:
x1 = [1,2,3,4,5]
x2 = [1,4,9,16,25]
x3 = [1,2,3,4,5]


# 1 subplot
plt.figure(figsize=(20,10))
# plt.grid()
# plt.show()

plt.subplot(2,2,1)
plt.plot(x1, x2, color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x1, x3, color='violet')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x1, x2, color='red')
plt.title("Plot 3")
```
