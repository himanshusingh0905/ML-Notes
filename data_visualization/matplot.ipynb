```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Let's customize above plots:

plt.plot(x,y, color='red', linestyle='--', marker='o',linewidth=5, markersize=15)
plt.grid()
plt.xlabel('X - axis') 
plt.ylabel('Y - axis')  
plt.show()
```

```python
# Multiple plots:
x1 = [1,2,3,4,5]
x2 = [1,4,9,16,25]
x3 = [1,2,3,4,5]


# 1 subplot
plt.figure(figsize=(20,10))
# plt.grid()
# plt.show()

plt.subplot(2,2,1)
plt.plot(x1, x2, color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x1, x3, color='violet')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x1, x2, color='red')
plt.title("Plot 3")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Start point
x0, y0, z0 = 1, 2, 3

# Direction vector
vx, vy, vz = 3, 4, 5

# Plot vector as a line (from t = -1 to 2)
t_vals = [-1, 0, 1, 2]
x = [x0 + vx*t for t in t_vals]
y = [y0 + vy*t for t in t_vals]
z = [z0 + vz*t for t in t_vals]

ax.plot(x, y, z, label='Direction Vector Line', color='blue')
ax.scatter(x0, y0, z0, color='red', label='Starting Point')
ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

for i in range(2):
    for j in range(2):
        axs[i, j].plot([1, 2, 3], [2 * (i + j + 1) * x for x in [1, 2, 3]])
        axs[i, j].set_title(f"Plot {i},{j}")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axs[i, j].plot([1, 2, 3], [2 * (i + j + 1) * x for x in [1, 2, 3]])
        axs[i, j].set_title(f"Plot {i},{j}")

plt.tight_layout()
plt.show()
```
