
## 2. Bivariate Analysis

### Definition
**Examines the relationship between two variables**

### Purpose
- Determine if variables are related
- Quantify the strength and direction of relationships
- Identify patterns or trends between variables
- Test hypotheses about variable relationships

### Common Techniques

#### Statistical Measures
- **Correlation Coefficient**: Pearson's r, Spearman's rank, Kendall's tau
- **Covariance**: Measures how two variables change together
- **Chi-Square Test**: Tests association between categorical variables
- **t-Test**: Compares means between two groups
- **Simple Linear Regression**: Models linear relationship between variables

#### Visualizations
- **Scatter Plot**: Shows relationship between two numerical variables
- **Line Plot**: Shows relationship with time as one variable
- **Grouped Bar Chart**: Compares categories across two variables
- **Heatmap**: Shows correlation or cross-tabulation
- **Box Plot (grouped)**: Shows distribution of one variable across categories of another

## Scatter Plot: 
- Relationship between two **numerical** variables.

#### `plt.scatter()` or `sns.scatterplot()`

```python
""" 
There can be many relationships we can examine:
  1. Fare and Age
  2. Sex and Fare
  3. passenger_class and Fare
  4. survived vs Sex 
  etc.
"""

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
sns.scatterplot(x=df['Age'], y=df['Fare'], color='yellow')


plt.subplot(1,2,2)
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)   # use of hue, third variable is using 'hue' for visualization.
sns.regplot(x ='')

print(" survived count : ", df['Survived'].sum())
```
